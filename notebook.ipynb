```python
import pdfplumber

import warnings
warnings.filterwarnings("ignore", message="CropBox missing.*")

parsed = []

try:
    with pdfplumber.open("sample/statements/maybank.pdf") as pdf:
        for page in pdf.pages:
            tables = page.extract_tables()
            for table in tables:
                for row in table[1:]:
                    parsed.append(row)
except:
    print()

parsed
```

```python
import pandas as pd

rows = []

for page in parsed:
    desc_lines = page[1].strip().split("\n")[1:]
    print(desc_lines)
    amt_lines = page[2].strip().split("\n")

    balance = float(amt_lines[0])
    amt_lines = amt_lines[1:]

    
    
    desc_block = []
    amt_idx = 0
    stop_words = ("BOOSTER", "DUITNOW", "Fund Transfer", "QR", "MBB CT", "MAE QR", "BANKI")
    
    for line in desc_lines:
        if amt_idx >= len(amt_lines):
            break
        desc_block.append(line)
        if len(desc_block) >= 3 and desc_block[-1].lstrip().startswith(stop_words):
            full_desc = " ".join(desc_block).strip()
            amt = amt_lines[amt_idx]
            balance += float(amt[:-1])
            rows.append(["", full_desc, amt, balance])
            amt_idx += 1
            desc_block = []

df = pd.DataFrame(rows, columns=["Date", "Description", "Amount", "Balance"])

display(df)
```
